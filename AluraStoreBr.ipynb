### Importação dos dados



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
loja.info()
```

```python
loja.shape
```

```python
loja.describe()
```

```python
faturamentoLojaUm =  loja['Preço'].sum()
faturamentoLojaDois = loja2['Preço'].sum()
faturamentoLojaTres = loja3['Preço'].sum()
faturamentoLojaQuatro = loja4['Preço'].sum()

print(f'Faturamento Loja 1: {faturamentoLojaUm:.2f}')
print(f'Faturamento Loja 2: {faturamentoLojaDois:.2f}')
print(f'Faturamento Loja 3: {faturamentoLojaTres:.2f}')
print(f'Faturamento Loja 4: {faturamentoLojaQuatro:.2f}')
```

```python
grBarra = [faturamentoLojaUm, faturamentoLojaDois, faturamentoLojaTres, faturamentoLojaQuatro]
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

plt.figure(figsize=(15, 5))
plt.bar(lojas, grBarra)

plt.title('Faturamento Total por Loja', fontsize=14)
plt.xlabel('Lojas', fontsize=12)
plt.ylabel('Faturamento (R$)', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

plt.figure(figsize=(15, 5))
plt.pie(grBarra, labels=lojas, autopct='%1.1f%%')

plt.title('Distribuição do Faturamento por Loja', fontsize=16)

plt.axis('equal')
plt.show()
```

# 2. Vendas por Categoria


```python
loja.head()
```

```python
def quebraLinha():
  print("\n")
  print("=" * 50)
  print("\n")
```

```python
vendas_por_categoria_loja1 = loja.groupby('Categoria do Produto')['Preço'].sum()
vendas_por_categoria_loja2 = loja2.groupby('Categoria do Produto')['Preço'].sum()
vendas_por_categoria_loja3 = loja3.groupby('Categoria do Produto')['Preço'].sum()
vendas_por_categoria_loja4 = loja4.groupby('Categoria do Produto')['Preço'].sum()

quebraLinha()
print(f'Loja 1 : {vendas_por_categoria_loja1}')
quebraLinha()
print(f'Loja 2 : {vendas_por_categoria_loja2}')
quebraLinha()
print(f'Loja 3 : {vendas_por_categoria_loja3}')
quebraLinha()
print(f'Loja 4 : {vendas_por_categoria_loja4}')

```

# 3. Média de Avaliação das Lojas

```python
media_loja1 = loja['Avaliação da compra'].mean()
media_loja2 = loja2['Avaliação da compra'].mean()
media_loja3 = loja3['Avaliação da compra'].mean()
media_loja4 = loja4['Avaliação da compra'].mean()

print(f'Média Avaliação Loja 1: {media_loja1:.2f}')
print(f'Média Avaliação Loja 2: {media_loja2:.2f}')
print(f'Média Avaliação Loja 3: {media_loja3:.2f}')
print(f'Média Avaliação Loja 4: {media_loja4:.2f}')
```

```python
lojas = ['Loja 1', 'Loja 2', 'Loja 3','loja 4']
media_lojas = [media_loja1, media_loja2, media_loja3, media_loja4]

plt.figure(figsize=(15, 5))
plt.plot(lojas, media_lojas, marker='o', linestyle='-', color='b', linewidth=2, markersize=8)

plt.title('Média de Avaliação por Loja', fontsize=14)
plt.xlabel('Lojas', fontsize=12)
plt.ylabel('Média de Avaliação', fontsize=12)

for i, v in enumerate(media_lojas):
    plt.text(i, v , f'{v:,.2f}', ha='center', va='bottom', fontsize=10)
# Mostrar o gráfico
plt.tight_layout()
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
loja.head()
```

```python
produto_mais_vendido = loja.groupby('Produto')['Preço'].sum().idxmax()
produto_menos_vendido = loja.groupby('Produto')['Preço'].sum().idxmin()

produto_mais_vendido_loja2 = loja2.groupby('Produto')['Preço'].sum().idxmax()
produto_menos_vendido_loja2 = loja2.groupby('Produto')['Preço'].sum().idxmin()

produto_mais_vendido_loja3 = loja3.groupby('Produto')['Preço'].sum().idxmax()
produto_menos_vendido_loja3 = loja3.groupby('Produto')['Preço'].sum().idxmin()

produto_mais_vendido_loja4 = loja4.groupby('Produto')['Preço'].sum().idxmax()
produto_menos_vendido_loja4 = loja4.groupby('Produto')['Preço'].sum().idxmin()

quebraLinha()
print('Loja 1:')
print(f'Produto mais vendido: {produto_mais_vendido}')
print(f'Produto menos vendido: {produto_menos_vendido}')

quebraLinha()
print('Loja 2:')
print(f'Produto mais vendido: {produto_mais_vendido_loja2}')
print(f'Produto menos vendido: {produto_menos_vendido_loja2}')

quebraLinha()
print('Loja 3:')
print(f'Produto mais vendido: {produto_mais_vendido_loja3}')
print(f'Produto menos vendido: {produto_menos_vendido_loja3}')

quebraLinha()
print('Loja 4:')
print(f'Produto mais vendido: {produto_mais_vendido_loja4}')
print(f'Produto menos vendido: {produto_menos_vendido_loja4}')

```

# 5. Frete Médio por Loja


```python
loja.head()
```

```python
media_frete_loja1 = loja['Frete'].mean()
media_frete_loja2 = loja2['Frete'].mean()
media_frete_loja3 = loja3['Frete'].mean()
media_frete_loja4 = loja4['Frete'].mean()

quebraLinha()
print(f'Frete Médio Loja 1: {media_frete_loja1:.2f}')
quebraLinha()
print(f'Frete Médio Loja 2: {media_frete_loja2:.2f}')
quebraLinha()
print(f'Frete Médio Loja 3: {media_frete_loja3:.2f}')
quebraLinha()
print(f'Frete Médio Loja 4: {media_frete_loja4:.2f}')
```

```python
fretes = [media_frete_loja1, media_frete_loja2, media_frete_loja3, media_frete_loja4]

lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

cores = ['#FF6347', '#4682B4', '#32CD32', '#FFD700']

plt.figure(figsize=(15, 5))
plt.bar(lojas, fretes, color=cores)

plt.title('Média de Frete por Loja', fontsize=14)
plt.xlabel('Lojas', fontsize=12)
plt.ylabel('Média de Frete (R$)', fontsize=12)

for i, v in enumerate(fretes):
    plt.text(i, v , f'R$ {v:,.2f}', ha='center', va='bottom', fontsize=10)

# Mostrar o gráfico
plt.tight_layout()
plt.show()
```
